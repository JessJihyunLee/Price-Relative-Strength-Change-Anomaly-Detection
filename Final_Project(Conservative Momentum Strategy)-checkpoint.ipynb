## Strategy 1. Using average momentum

  
- Strategies  
    1. Calculate returns of previous 6 months
    2. Compare with SHY
    3. If the return of ETF is better than SHY, select the ETF 
    
    

- constraint
    1. Each ETF can have weight 0 or 1/(number of ETFs)
    2. Weight of SHY : 1 - sum of weights of ETFs

```python
def calc_avg_mom(ret, periods=[3,6,9,12]):
    res = pd.DataFrame(0.0, index=ret.index, columns=ret.columns)
    
    for i in periods:
        monthly_ret = ret.rollapply(i, lambda x : (x.add(1).prod()**(1.0/i)) - 1)
        res += monthly_ret
    
    return(res.dropna()/len(periods))
```

```python
def calc_ret_pf1(avg_mom, ret_m, asset_number = 3):
    dt_join = avg_mom.join(ret_m['shy'])
    dt_join = dt_join.sub(dt_join.shy, axis=0)
    
    dt_weight = dt_join.apply(lambda x : x[x >= 0].nlargest(asset_number), axis=1)
    dt_weight[dt_weight > 0] = 1/asset_number
    dt_weight[dt_weight.isna()] = 0
    dt_weight.shy = 1 - dt_weight.sum(axis=1)

    pf_ret = pd.DataFrame([], index=dt_join.index[1:])
    pf_ret['ret'] = (dt_weight.shift(1) * ret_m).sum(axis=1)
    
    return(pf_ret)
```

```python
etf_avg_mom = calc_avg_mom(etf_ret_m, periods=[6])
etf_avg_mom.head()
```

```python
ret_pf1 = calc_ret_pf1(etf_avg_mom, ret_m, asset_number=1)
ret_pf2 = calc_ret_pf1(etf_avg_mom , ret_m, asset_number=2)
ret_pf3 = calc_ret_pf1(etf_avg_mom , ret_m, asset_number=3)
ret_pf4 = calc_ret_pf1(etf_avg_mom , ret_m, asset_number=4)
ret_pf5 = calc_ret_pf1(etf_avg_mom , ret_m, asset_number=5)
ret_pf6 = calc_ret_pf1(etf_avg_mom , ret_m, asset_number=6)
```

## Strategy2. Using average momentum with RSI
<br/>
* using RSI as ETF selection criteria to exclude ETFS that might drop drastically <br/>
* We set threshold in two different ways :<br/>
     - RSI lower than 70. Because RSI over 70 is often considered as overbought<br/>
     - Exclude high outliers of RSI. Instead of setting fixed criteria for every period, we used Local Outlier Factor to exclude based on density of that time.

```python
def calc_financial_dat(data, stat):
    res_dt = pd.DataFrame([], index=data_m.index)
    assets = list(data.columns[data.columns != 'shy'])
    for asset in assets:
        dt = data[[asset]]
        dt.columns = ['close']

        s_dt = Sdf.retype(dt)
        s_dt = s_dt[[stat]].resample('M').agg('last')
        s_dt.columns = [asset]

        res_dt = pd.concat([res_dt, s_dt],axis=1)
    return(res_dt)
```

### 2-2. Selecting ETFs without high RSI outliers (outlier detection using One Class Support Vector Machine)

* Based on whether market is up trend (bull market) or down trend (bear market), RSI cutting threshold to decide whether its oversold or overbought or not needs to be different.
[Reference]('https://www.fidelity.com/learning-center/trading-investing/technical-analysis/technical-indicator-guide/RSI')
* One Class Support Vector Machine

```python
#Get Daily RSI
def RSIfun(price, n=14):
    delta = price.diff()
    dUp, dDown = delta.copy(), delta.copy()
    dUp[dUp < 0] = 0
    dDown[dDown > 0] = 0
    RolUp = dUp.rolling(n).mean()
    RolDown = dDown.rolling(n).mean().abs()
    RS = RolUp / RolDown
    rsi= 100.0 - (100.0 / (1.0 + RS))
    return rsi
```

## Startegy 3. Using average momentum with DMI (Dynamic Momentum Index)

```python
#DMI : limited calculated time period from DMI to be between 5 and 30
def get_time(price):
    #Value available from 14th period
    std5=price.rolling(5).std()
    stda=std5.rolling(10).mean()
    v=std5/stda
    t=14/v.round()
    t[t.isna()]=0
    t=t.map(lambda x : int(min(max(x,5),30)))
    return t
```

```python
def DMIfun(index,stock_name):
    time=t[stock_name].iloc[index]
    if (index-time)< 0 :
        subset=data[stock_name].iloc[0:index]
    else:
        subset=data[stock_name].iloc[(index-time):index]
    delta = subset.diff()
    dUp, dDown = delta.copy(), delta.copy()
    dUp[dUp < 0] = 0
    dDown[dDown > 0] = 0
    avgUp=dUp.mean()
    avgDown=np.abs(dDown.mean())
    RS = avgUp/avgDown
    rsi=100.0 - (100.0/(1.0 + RS))
    return rsi
```

Get dataframe 'dmi' which has Dynamic Momentum Index for each stock as values 

```python
dates=Series(data.index)
periods=Series(range(14,len(dates)),index=data.index[14:].values)
assets = list(data.columns[data.columns != 'shy'])
t=data.apply(get_time)
dmi=DataFrame(0, columns=assets, index=periods.index)
for stock_name in assets:
    dmi[stock_name]=periods.map(lambda x : DMIfun(x,stock_name))
```
